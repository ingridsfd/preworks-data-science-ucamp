# Análisis Exploratorio de Datos

El Análisis Exploratorio de Datos o EDA (Exploratory Data Analysis, por sus siglas en inglés) es la tarea más importante para un Científico de Datos, pues constituye el primer paso antes de preparar algún modelo.

Cuando se recibe la información por primera vez, es importante comprender el conjunto de datos, así como reconocer patrones y distribuciones que puedan ser útiles para entender el comportamiento de las variables e identificar anomalías que puedan afectar al momento de realizar el modelo.

El análisis exploratorio es un aspecto del científico de datos que está en mejora continua. Si bien algunos se aprenden con observación y práctica, existen determinadas herramientas o gráficos que indican distintas características.

- ¿Cuántos registros (o filas con datos) hay?

- ¿Existen datos raros o atípicos en el conjunto de datos?

## Completitud

La completitud mide cuánta información se tiene sobre una variable; en otras palabras, indica los datos que están completos en el conjunto de datos. Para saberlo, se realiza una tabla con una nueva variable. Su aplicación en el ejemplo `df` sería como sigue:

Se define una variable denomiada `completitud`, semejante al conjunto de datos `df`. En este caso, se utilizará la función `.sum()` para sumar los elementos que están vacíos (`.isnull()`).
Al juntar estas tres caractarísticas se obtiene `df.isnull().sum()` y dado que se pretende hacer un _Data Frame_ ,entonces se usará `pd.DataFrame`.

```python
#Verificamos completitud de las variables , es decir,  si tenemos valores nulos
completitud = pd.DataFrame(df.isnull().sum())
completitud
```

Como resultado, para cada variable (o columna) de `df` se determina cuántos valores distintos de cero tiene. De esta manera, a mayor cantidad mostrada, mayor número de datos faltantes.

De lo anterior, resultará un data frame donde cada «fila» constituye una de las columnas del `df` original; como se busca que la primera fila sea la `0`, la segunda `1` y así sucesivamente, entonces se usará `reset_index`para resetear los índices. El argumento `inplace= True` sirve para evitar la repetición de `completitud =completitud.reset_index()`, aunque también es válido.

```python
completitud.reset_index(inplace = True)
completitud
```

Como notaste en la tabla anterior, el índice comienza en 0 y avanza de uno en uno. Recuerda que en Python el conteo siempre inicia a partir de cero.

Probablemente, también notaste que este conjunto de datos **No tiene encebezados**; por ello, es necesario crearlos con `.rename(columns=`.

En este punto, **diccionarios** señala que `"index":"columna"`. Esto significa que la columna de los **índices** ahora se llamará «columna». Por lo anterior, `0:"total"` quiere decir que la columna `0` se llamára "total" (dado que en Python el conteo empieza desde 0 y los índices no valen).

```python
completitud = completitud.rename(columns = {"index":"columna",0:"total"})
completitud
```

Se creará una nueva columna llamada "completitud", igual a 1 - la división de la columna "total" entre la cantidad de columnas del conjunto de datos original. El resultado se multiplica por 100 para obtener un procentaje.

$$
Completitud_i = 1 - \frac{total_i}{{total\ columnas}} \times 100
$$

```python
#completitud["completitud"] = (1 - completitud["total"] / df.shape[0] * 100)
completitud['completitud'] = ((1 - completitud['total'] / df.shape[0]) * 100).round(2)
completitud
```

Por último, se organiza de mayor a menor con `.sort_values` (no olvides indicar `ascending=True` para obtener el orden correcto) y se resetean nuevamente los índices.

```python
#completitud.sort_values( by = 'completitud')
completitud = completitud.sort_values(by = "completitud", ascending = True)
#completitud.reset_index(drop = True, inplace = True)
completitud
```

Con todos los pasos anteriores se obtendrá una tabla como la mostrada arriba. Esto significa que en el conjunto de datos (`df`), "fila" se traduce en cada variable (o columna). Además, dicha tabla indica cuántos valores «nulos» tiene, es decir, las celdas vacías por cada columna. Por último, se muestra la «completitud» de un número que, mientras más negativo es, aportará menos información.

Antes de avanzar con este Bootcamp, necesitas conocer algunos conceptos básicos.

Exiten tres distintos tipos de variables, dos de ellas hacen referencia a las características de los números que la integran y la última alude a la variable buscada.

Dichas variables aparecen en toda la informción, más aún en los conjuntos de datos.

Es altamente aconsejable que antes de cualquier análisis explores un poco cada variable dentro de un conjunto de datos y determines de qué tipo son, pues esto ayudará más adelante para hacer un diagnóstico más cercero y diseñar mejores modelos.

Los tipos de variables son:

- Variables continuas: sus elementos son números que no pueden contarse de uno en uno, es decir, no son categorizables. Ejemplo de esto es la estatura, pues no existen personas de 1 metro e inmediatamente la medida salta a 2 metros, sino que hay toda una gama infinita de números que podrían verse en un rango. Otro ejemplo son los litros, mediante los cuales se miden valores con muchos decimales; por ejemplo, si se pide un valor para esta variable, no sería posible limitarlo a pocas respuestas. Es como señalar en una regla un valor; si lo piensas detenidamente, las respuestas son muchísimas.

- Variables discretas: estas pueden categorizarse. El primer ejemplo son los valores "cerrados", como el número de casas compradas o reportes emitidos; no puede contarse media casa o medio reporte. Otro ejemplo muy usual son los colores; algo puede ser "rojo", "negro" o "azul", pues no hay valores intermedios.

- Variable objetivo: es aquella que desea predecirse mediante la implementación de un modelo de predicción.

Para comprender mejor el tipo de efecto de cada variable (o columna), se modificará el nombre para agregar lo siguiente al incio del de cada columna:
- «v_» si es una variable discreta.
- «c_» si es una variable continua.
- «tgt_» si es la variable objetivo.

Recordarás que utilizaste el método `.rename`, pero el mayor obstáculo sería repetir el mismo método.

Ahora, utiliza las funciones de **Python** para hacer una **función** que denominarás `rename_cols` y tomará tres parámetros diferentes:
1. `df` el conjunto de datos del cual tomará la información.
2. `cols` para todas las columnas a las cuales les modificarás el nombre.
3. `prefix` el prefijo que asignarás al nombre de cada columna.

Si bien puedes nombrarlos en español, es recomendable hacerlo en inglés, dado que permite que tus programas tengan un mayor alcance con la comunidad mundial.

Para hacerlo, verás el paso a paso para definir una función.
1. Indica el nombre de la función.
2. Crea una nueva variable llamada `new_feats`, que será igual al prefijo indicado e indica el nombre de cada columna dentro de la lista delimitada en `cols`.
3. Define una nueva variable llamada `df`, que usará de manera simultánea el `.rename`.
4. Recuerda usar un `return` para que la función use el nuevo `df` que realizaste.

```python
#Creamos una función que nos permita modificar el nombre de nuestras columnas
def rename_cols(df,cols,prefix):
  new_feats=[prefix+col for col in cols]
  df=df.rename(columns=dict(zip(cols,new_feats)))
  return df
```

Como verás, ya tienes tres variables nuevas que incluyen el nombre de todas las columnas y están asignadas, ya sea a `v_feats`, `c_feats` o `tgt`.

# Análisis Univariable

Luego de detectar los tipos de variables presentes en el conjunto de datos, es posible identificar una en especial: la **variable objetivo**.

Al identificar esta **variable objetivo**, se aplica el **análisis univariable**, basado en estudiar la **variable objetivo** y su **relación con las demás variables** para encontrar correlaciones o algún dato importante relacionado con esta variable.

Una forma efectiva de comprender los valores estadísticos de una o más variables es `.describe()`. Este método permite conocer:
- `count`: la cantidad de elementos contados en la variable.
- `mean`: valor promedio de los elementos dentro de la variable.
- `std`: la variación estándar de los elementos dentro de la variable.
- `min`: el elemento más pequeño dentro de la variable.
- `25%`: el valor del cuartil 1 o percentil 25.
- `50%`: el valor del cuartil 2 o percentil 50.
- `75%`: el valor del cuartil 3 o percentil 75.
- `max`: el elemento más grande dentro de la variable.

```python
#La sentencia "describe" nos indica la estadistica
df.filter(like="tgt_").describe()
```

```python
sns.distplot(df['tgt_SalePrice']);
```

La gráfica de arriba es una **histograma** e indica cómo se comportan todos los números que están en la **variable objetivo**, llamado `tgt_SalePrice`.

Se observa que la mayoría de los resultados está por debajo de 200 000 pero por encima de 100 000 y que los valores arriba de 300 000 son muy pocos.

Este tipo de comportamiento se llama **distribución normal**,es decir, la mayoría de los números se concentra en un punto específico y se disipa a lo largo de la recta.

- Asimetría
Medida de distribución que puede tomar valores positivos o negativos. En caso de ser positivo, indica que la distribución se inclina hacia los valores positivos; en caso contrario, los valores negativos muestran que la distribución se inclina hacia los valores negativos.

- Curtosis
La curtosis es una medida estadística que determina el grado de acumulación en los valores de una variable alrededor del centro de la distribución. Cuando tiende a 3, significa que sigue una distribución normal, es decir, que los valores no son atípicos.

```python
print("Asimetría: %f" % df['tgt_SalePrice'].skew())
print("Curtosis: %f" % df['tgt_SalePrice'].kurt())
```

# Análisis multivariante

Anteriormente, conociste el "análisis univariable", que sucede entre cada variable con la variable objetivo. Sin embargo, para comparar múltiples variables contra otras tantas (más allá del objetivo), entonces se efectúa un "análisis multivariable".

Este análisis permitirá entender la relación de las variables entre sí para detectar anomalías o patrones internos.

```python
#Matriz de correlación general entre todas las variables
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

El mapa de calor permite el conocimiento visual de las variables y sus relaciones; cuando están muy correlacionadas podría indicar multicolinealidad, esto es, que tienen la misma información y no serían útiles para nuestro modelo.

Esta gráfica proveniene del "análisis univariable", en particular, del estudio de la correlación entre cada una de las variables y la variable objetivo; sin embargo, en este otro caso se comparan las correlaciones de **todas las variables** contra **todas las variables**.

Esto genera una matriz para señalar los valores que, de entre los distintos factores dentro un conjunto de datos, se comportan de manera similar o no tienen relación alguna.

Aunque para este momento no resulte del todo intelegible, en cuanto avances con los modelos, estos cobrarán sentido e importancia.

```python
#En vista de nuestras variables más correlacionadas graficaremos un "Scatter plot" para facilitar la comprensión de nuestras variables
sns.set()
cols = ['tgt_SalePrice', 'v_OverallQual', 'c_GrLivArea', 'c_GarageCars', 'c_TotalBsmtSF', 'v_FullBath']
sns.pairplot(df[cols], size = 2.5)
plt.show();
```

La gráfica anterior se asemeja al mapa de calor previo, la diferencia radica en que muestra pequeñas gráficas de la relación de **todas las variables** contra **todas las variables**, así, se entiende como la forma "gráfica" de corroborar si existe correlación.

Recuerda que si observas que una línea se dibuja quiere decir que los valores de esas variables poseen una alta correlación.
