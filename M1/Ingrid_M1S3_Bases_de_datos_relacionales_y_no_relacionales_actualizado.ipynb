## **3.1 ¿Qué es una base de datos?**

Una **base de datos** es un conjunto de información organizada de tal forma que pueda estar estructurada en un sistema informático, en otras palabras,una base de datos es una agrupación de datos pertenecientes a un mismo contexto alojada en un gestor de bases de datos.

Dentro de las bases de datos nos encontramos con diferentes modelos, tipos, composiciones, cada uno con ventajas y desventajas, por ejemplo acorde a su organización, jerarquía, relación con otras tablas, entre otras.


Algunos ejemplos de bases de datos son los registros transaccionales, es decir, las busquedas realizadas en Google, los logeos en una página, el registro de las compras en una tienda departamental, entre otras. Todas generan un conjunto de registros que va a parar a una base de datos alojada en algún gestor de bases de datos.

Al mantener esta información en una base de datos, nos permitirá hacer una consulta de esa información de manera estructurada, además de guardar un histórico de estos datos, una base de datos puede alojar cualquier información con lo que podemos tomar estrategías para presentar, consultar o hacer cualquier uso con dicha información.

## **3.2 Bases de datos relacionales**

Una vez plasmado el concepto de una base de datos, existen diferentes tipos de estas. Una base de datos relacional es un conjunto de relaciones formadas por columnas y registros, conformados por llaves primarias y foráneas, lo cual permite crear relaciones con otras tablas o bases de datos.


Una de las principales características de la base de datos relacional son las relaciones que se pueden crear por medio de las llaves, es decir, una tabla está conformada por una columna con una llave primaria (normalmente, algún identificador abreviado, como "ID") y columnas con llaves foráneas, lo cual permite conectar o relacionar otra tabla por medio de esas llaves, por lo que una columna con una llave primaria se relacionará con otra tabla con una columna con una llave foranéa que tendrán en común para poder extraer información de ambas tablas con dicha relación.


La base de datos está dividida en dos secciones:
- El esquema: guarda la información del conjunto de tablas junto con sus relaciones, formación de columnas y  registros.
- Las tablas: son el conjunto de información agrupada en columnas y registros.

## **3.3 Motores de bases de datos**


Un motor de base de datos es un software que funciona como un sistema de administración de la base de datos, el cual permite crear esquemas, bases de datos y tablas, así como leer, actualizar, modificar y eliminar datos en algún esquema o en alguna base de datos; en otras palabras, un motor de base de datos permite administrar y gestionar las bases de datos, incluidos los permisos, usuarios o servicios que este puede generar.

Un motor de base de datos se compone de tres estructuras:

- Lenguaje de definición de datos: se utiliza para definir la estructura de la base de datos.
- Lenguaje de manipulación de datos: se utiliza gestionar los datos en un esquema o una base de datos.
- Lenguaje de control de datos: se utiliza para administrar los permisos y el acceso a un esquema o una base de datos.


Los motores de bases de datos deben cumplir con las siguientes características:

- Abstracción de la información.
- Independencia.
- Redundancia mínima.
- Consistencia.
- Seguridad.
- Integridad.
- Respaldo y recuperación.
- Control de la concurrencia.
- Tiempo corto de respuesta.

Existe una gran cantidad de proveedores que ofrecen el servicio de gestión de bases de datos; entre los más populares están SQL Server, MySQL, Oracle e IBM.

## **3.4 Relaciones en bases de datos**

Las relaciones de bases de datos son asociaciones entre tablas que se crean utilizando sentencias de unión para recuperar datos de 2 o más tablas.

Existen varias formas de relacionar bases de datos, es decir, juntar información por medio de componentes denominados *llave primaria* y *llave foránea*, mediante los cuales puedes relacionar una tabla con otra.

Existen tipos de uniones en las tablas:

- **Inner join:** implica que si estás uniendo dos tablas en una columna en común, solo retornarán los datos que coincidan en ambas tablas.
- **Left join:** en este tipo de unión solo retornan todos los datos de la tabla del lado izquierdo, si los datos coinciden con la tabla de la mano derecha.
- **Right join:** este tipo de unión es opuesto al anterior, pues implica que solo retornarán los datos de la tabla del lado derecho, si los datos coinciden con la tabla del lado izquierdo.
- **Outer join:** implica que si estás uniendo dos tablas en una columna en común, retornarán los datos de ambas tablas, con lo cual dejan nulos las columnas que no coincidan.

En **Python**, puedes hacer uso de estos tipos de uniones para relacionar 2 o más *data frames*. La función "pandas.merge" permite realizar *joins* entre tablas. El join es realizado en las columnas o en las filas. En el primer caso, las etiquetas de las filas son ignoradas. En cualquier otro caso (*joins* realizados entre etiquetas de filas o entre etiquetas de filas y de columnas), las etiquetas de las filas se mantienen.

A continuación se presenta un primer ejemplo.
Se parte de dos tablas que contienen las ventas y los costos de producción para varios meses:

```python
#Creamos un data frame con dos columnas, Mes y Ventas, con datos simulados
import pandas as pd
df=pd.DataFrame({"Mes": ["ene","feb","mar","may"],
                "Ventas": [15,10,12,11]})
df
```

```python
df2=pd.DataFrame({"Mes": ["feb","ene","mar","abr"],
                "Costos": [5,8,4,7]})
df2
```

Observa que ambos *data frames* tienen una columna en común ("Mes") y varias filas comunes ("ene", "feb" y "mar").
Observamos que en df2, las filas no están ordenadas y que en df, el mes de enero tiene como índice 0, mientras que en df2 el mes de enero tiene como índice 1.

Si aplicas la función merge a estos *data frames* con los valores que aparecen de manera automática, obtendrás el siguiente resultado:



```python
pd.merge(df,df2) # Duda: ¿No importa si los datos están desordenados en "Mes", podemos continuar aplicando la función Merge?
#Pregunta ¿qué tipo de unión se aplicó en este caso?
```

Esos valores automáticos suponen que el *join* se realiza en las columnas comunes y tipo *inner* (considerando solo las filas con etiquetas comunes).

Si especificas que el *join* es tipo *outer*, lo cual se define con el parámetro "how", el resultado considerará todas las etiquetas presentes en ambos *data frames*:

```python
pd.merge(df,df2,how='outer')
```

Como puedes observar, se rellenaron con "NaN's" los valores inexistentes. Otras opciones para el parámetro "how" son "left" y "right".


```python
pd.merge(df,df2,how='left')
```

```python
pd.merge(df,df2,how='right')
```

Ya se ha comentó que, de manera automática, el *join* se realiza entre las columnas comunes. Sin embargo, es controlable con el parámetro "on" y especificando la columna o las columnas a usar. Por ejemplo, considera los siguientes *data frames*:

```python
df=pd.DataFrame({"Mes": ["ene","ene","feb","feb"],
                  "Producto":["A","B","A","B"],
                "Ventas": [10,15,6,11]})
df
```

```python
df2=pd.DataFrame({"Mes": ["ene","ene","feb","feb"],
                  "Producto":["A","B","A","B"],
                "Costos": [5,8,4,7]})
df2
```

Hay dos columnas comunes, lo cual supone que el resultado de un "merge", de manera automática, sería el siguiente:

```python
pd.merge(df,df2)
```

Es decir, en cada combinación de mes-producto se añadirían los valores de los campos de ventas y costos. Si quisieras que el *join* se realizara solo en uno de los campos, por ejemplo, en Product, bastaría con especificarlo con el parámetro "on":

```python
pd.merge(df,df2,on="Producto")
```

Además del campo utilizado para realizar el *join* (columna:producto), al existir un campo común en ambos *data frames* (Mes) que no se desea usar para el *join*, pandas añade un sufijo  a este campo en ambas tablas para diferenciarlo.

También podría ocurrir que ambos *data frames* no tengan columnas comunes (columnas con el mismo nombre), pero, aun así, quisieras realizar el *join* en algunas de ellas, por ejemplo:

```python
df=pd.DataFrame({"Mes": ["ene","feb","mar","may"],
                "Ventas": [10,15,6,11]})
df
```

```python
df2=pd.DataFrame({"Mes_nombre": ["feb","ene","mar","abr"],
                "Costos": [5,8,4,7]})
df2
```

Al no haber columnas comunes, la ejecución de la función **merge** arrojaría un error. En este caso se pueden usar los parámetros **left_on** y **right_on** para especificar el campo a usar en la tabla de la izquierda del **join** y en la de la derecha, respectivamente:

```python
pd.merge(df,df2, left_on='Mes',right_on='Mes_nombre')
```

## **3.5 Bases de datos no relacionales**

Las bases de datos no relacionales son un sistema de almacenamiento de información que se caracteriza por no usar el lenguaje SQL para las consultas.

Otra de sus principales características es que no trabajan con estructuras definidas. Es decir, los datos no se almacenan en tablas relacionales, por lo que la información no se registra en columnas y registros como lo observamos en las bases de datos relacionales

Entre sus ventajas, tienen una gran escalabilidad y están pensadas para la gestión de grandes volúmenes de datos.

Las bases de datos no relacionales son más actuales que las relacionales, y su desarrollo se ha basado en la necesidad de crear sistemas de gestión capaces de trabajar con datos no estructurados pero con una gran volumetría.
Algunas de sus principales ventajas son:
- Son mucho más flexibles a la hora de crear esquemas de información, lo que las convierte en una solución ideal para el almacenamiento y gestión de datos no estructurados.
- Pueden soportar mayores volúmenes de datos y añadir mayor capacidad.
- Garantizan un alto rendimiento, ya que están diseñadas para trabajar con modelos de datos concretos.

Por ejemplo, las busquedas realizadas por el buscador de "Google" son almacenadas en una base de datos no relacional ya que al ser cantidades gigantes de datos, este tipo de base de datos es ideal para almacenar este tipo de datos.

M1S3_Imagenes_Data Science (6).jpg

Entre sus desventajas, están las siguientes:
- No cumplen con las mismas propiedades de consistencia, integridad y durabilidad como lo hacen las bases de datos relacionales-
- No son compatibles con consultas en lenguaje SQL.
- Carecen de un sistema estandarizado, ya que todavía son bases de datos relativamente nuevas.
- Al ser un sistema de gestión de bases nuevos, no contamos con mucho soporte en dudas que pudieran surgir y su implementación suele ser más complicada.


M1S3_Imagenes_Data Science (7).jpg

Existen diferentes tipos de bases de datos no relacionales, en función del método que emplean para almacenar la información.

- **CLAVE-VALOR:** Se trata de bases de datos no relacionales que almacenan la información en base a pares de clave valor. Es decir, cada clave sirve como un identificador único, y a cada una de ellas se le aplica un valor.

- **GRÁFICOS:** Las bases de datos no relacionales basadas en gráficos están pensadas para crear relaciones y navegar por ellas. Las entidades de datos se almacenan mediante nodos y los bordes son los que crear las relaciones entre entidades. Con frecuencia las bases de datos gráficas se emplean en redes sociales, sistemas de detección o prevención de fraudes o sistemas de recomendaciones.
- **EN MEMORIA:** Son bases de datos no relacionales diseñadas para ofrecer respuestas en milisegundos y soportar grandes picos de tráfico. Un ejemplo de bases de datos en memoria son las empleadas en tablas de clasificaciones de juegos o en herramientas para hacer análisis en tiempo real, por su rápido acceso


M1S3_Imagenes_Data Science (8).jpg

Principalmente, las bases de datos no relacionales se emplean para almacenar datos no estructurados, o de los que no se tienen ningún esquema o información predefinida.
Esto hace que las bases de datos no relacionales sean una excelente opción para casos de uso de baja latencia y alto rendimiento, es decir, que necesitan ser escalables sin una limitante como lo tendríamos en una base de datos relacional.
