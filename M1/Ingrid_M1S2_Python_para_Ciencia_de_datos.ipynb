#### Array de un rango

#### **Promedio**

Pero antes de revisar esa función, se repasará lo que es el **promedio** o la **media**.

Imagina que tienes el siguiente **array**, que representa las calificaciones de un grupo escolar.

````python
grupo_c = np.array([10, 9, 7, 8, 9, 10, 6, 7, 8, 10, 8, 6, 5, 6, 9, 8, 7, 10, 9, 6])
````

El promedio no es más que la suma de todos los valores entre la cantidad de elementos que existen. De esa manera puedes sacar un número que representaría la calificación que tendría el grupo en general:
$$
promedio= \bar{x}=\frac{\text{suma de todas las calificaciones}}{\text{cantidad de calificaciones}}= \frac{\sum_{i=1}^{n}{calificaciones}}{n}
$$
En el ejemplo, si sumas todos los elementos, daría $158$, por lo que debes dividirlo entre la cantidad de calificaciones, que es $20$. Así, la fórmula sería:
$$
promedio= \bar{x}= \frac{158}{20}=7.9
$$
No solo existe una línea de código que te ayuda a encontrar ese número, sino dos:

````python
grupo_c.mean()
````

````python
np.mean(grupo_c)
````

#### **Mediana**

Si ordenaras todos los valores del **array** de menor a mayor, el que está en medio de toda la fila es conocido como la **mediana**:

median (1).png

> **NOTA**: en el ejemplo, al tratarse de un **array** de números pares, no hay un valor medio, sino 2: el que está en la décima posición y el que está en decimonónica primera posición. En esos casos, la media es el promedio de esos dos valores.

Para saberlo fácilmente, se escribe el siguiente código en **Python**:

````python
np.median(grupo_c)
````

## **2.3 Pandas**
**Pandas** es la evolución de **Numpy** pero con funciones avanzadas. Es una librería que permite el análisis y la manipulación de datos en forma de _data frames_. Pero ¿qué es un _data frame_? Se puede definir como una tabla de Excel en la que cada _data frame_ tiene encabezados, como los tiene una tabla de Excel, y cada elemento es identificado por un número de fila, que en **Pandas** es conocido como índice (si te lo estás preguntando, sí, igual que las **listas** de **Python** y que inician en `0`.)

Lo primero que aprenderás es cómo importarlo.

````python
import pandas as pd
````

### **Creación de data frames**

Existen muchas formas de importar los **data frames**. Aquí se mostrarán las dos formas más comunes de hacerlo.

La primera de ella es, directamente, desde el código que escribes.

Para que sea más sencillo, supón que quieres hacer un **data frame** similar al siguiente:

| Nombre | Edad | Estatura | ¿Beca? | Calificación |
| --- | --- | --- | --- | ---|
| Ana | 28 | 1.72 | Sí | 8 |
| Bruno | 33 | 1.69 | No | 7 |
| César | 34 | 1.77 | No | 8 |
| Diego | 23 | 1.79 | Sí | 10 |
| Erika | 25 | 1.69 | Sí | 7 |

Como verás, está conformado por cinco columnas, las cuales se llaman, respectivamente, Nombre, Edad, Estatura, ¿Beca? y Calificación. Además, puedes ver que consta de cinco filas (no se considera la fila de los encabezados), por lo que la primera, la cual tendrá el índice `0`, es la referente a Ana, y la fila con el número de índice `2` es la referente al tercer elemento: César.

Para poder capturar eso en **Python**, es necesario hacerlo de la siguiente manera:

````python
df = pd.DataFrame(
	{
    'Nombre': ['Ana', 'Bruno', 'César', 'Diego', 'Erika'],
		'Edad': [28, 33, 34, 23, 25],
		'Estatura': [1.72, 1.69, 1.77, 1.79, 1.69],
		'¿Beca?': [True, False, False, True, True],
		'Calificación': [8, 7, 8, 10, 7]
	}
)

df
````

Pon atención a la estructura:
- Primero se indica un nombre al **data frame**, como lo harías con una variable, por lo que puedes usar el nombre que más te agrade, seguido del signo `=`, ya que le asignarás un valor.
- La función es `pd.DataFrame`, seguido de un paréntesis de apertura: (`(`).
- A continuación sigue un **diccionario**, en el que cada clave es el nombre de cada columna. Cada columna está entre comillas (`'texto'`), pero también puedes usar comillas dobles (`"texto"`). Si indicas que se trata de una **cadena de texto**, entonces **Python** interpretará que es una variable y, por lo tanto, que ya definiste un valor para dicha variable antes de esta línea de código.
- Cada valor que tendrá toda la columna se indicará como una **lista**; por lo tanto, es importante decir que, para cada columna, el número de elementos de su respectiva **lista** deben ser el mismo. En el ejemplo, «Beca» tiene cinco elementos (todos son **booleanos**) y cada columna, a pesar de tener diferentes tipos de datos, tiene cinco elementos. Si alguno tuviera un dato de más o de menos, **Python** nos mostrará un error.
- Nota las comas (`,`) entre **lista** y **lista**.
- Nunca olvides que lo que abrió con paréntesis también debe cerrar con paréntesis (`)`).
> **NOTA**: cuidado con las mayúsculas y las minúsculas. Como pudiste observar, la función `pd.DataFrame` tiene una `D` y una `F` mayúsculas. Omitir esta recomendación generará un error en el código.

Existe un segundo método: que todos esos valores se guarden en variables o **listas**, y todo eso se junta a fin de crear el **data frame**.
> **NOTA**: no olvides que las listas son valores entre corchetes (`[` y `]`) y separados por comas (`,`).

````python
nombre_columnas = ['Nombre', 'Edad', 'Estatura', '¿Beca?', 'Calificación']

lista_ana = ['Ana', 28, 1.72, True, 8]
lista_bruno = ['Bruno', 33, 1.69, False, 7]
lista_cesar = ['César', 34, 1.77, False, 8]
lista_diego = ['Diego', 23, 1.79, True, 10]
lista_erika = ['Erika', 25, 1.69, True, 7]


df = pd.DataFrame(
	[lista_ana, lista_bruno, lista_cesar, lista_diego,	lista_erika],
	columns= nombre_columnas
)
df
````

Ahora analizarás lo que acabas de ver:
- Cada lista hace referencia a una fila. Como sucedió en el caso anterior, la cantidad de propiedades debe ser la misma.
- Cuando usas la función `pd.DataFrame` y abres un paréntesis (`(`), es inmediatamente precedido por una **lista** que incluye las **listas** con los valores de cada fila. En esta **lista** de **listas** no se incluye el nombre de los encabezados.
- Después de la **lista** de **listas** se debe colocar una coma (`,`), la cual indica que debes saltar a otro elemento, el cual se llama `columns` y va seguido de un signo `=` y después de la **lista** de los encabezados, que en este caso está en la variable `nombre_columnas`.
> **NOTA**: cuando veas el signo `=` en una función, no quiere decir que declararás una variable, como sucede cuando lo escribes en una línea independiente. Ese signo quiere decir que le asignarás un atributo. Esto puede parecer confuso al inicio, pero conforme vayas aprendiendo nuevas funciones estas se incorporarán naturalmente.

### **Filtros en un _data frame_**
Al igual que se pudo usar la función `.head()` en todo un **data frame**, también es posible usarla en una columna específica.

````python
df['Edad'].head(3)
````

Recuerda que cada fila del **data frame** tiene un índice para identificarla, donde la primero es `0`, la segunda es `1` y así sucesivamente. Ese índice sirve para hacer un «corte»; por ejemplo, si quieres conocer los datos de la segunda fila a la cuarta, debes hacer algo similar a lo siguiente.

````python
df[1:4]
````

El corte se hizo así porque el `1` que se anotó hace referencia a que iniciará en la segunda fila, y el `4` indica hasta dónde se detendrá, sin llegar a tocar el `4`, por lo que considerará como último valor el inmediato anterior, que es `3`, que para **Python** representa el cuarto valor.

Si quieres los valores de toda una fila, ingresas lo siguiente:

````python
df.loc[2] #Nos regresará todos los valores de la tercera fila.
````

Puedes usar los operadores lógicos para ayudarte a filtrar valores cumpliendo condiciones.

El siguiente código te ayuda a detectar en **data frame** todos los elementos que sacaron más de 7 de calificación:

````python
df[df['Calificación'] > 7]
````

O puedes indicar que cumplan una característica específica:

````python
df[df['¿Beca?'] == True]
````

Si quieres un filtro que cumpla dos condiciones, usarás el caracter `&` para indicar que debe plantear dos condiciones, las cuales deben cumplirse para mostrar la información.

````python
df[(df['Calificación'] > 7) & (df['¿Beca?'] == True)] #Duda: ¿Aquí por qué sigue mostrando 'Edad' y 'Estatura'?
````

#### **Modificar un _data frame_**

Hasta ahora los códigos que has visto sirven para explorar datos. Para modificar un **data frame**, debes usar el mismo método que se emplea para reescribir una variable.

El siguiente código te ayuda a reordenar la información de forma ascendente (de menor a mayor).

````python
df.sort_values('Calificación')
````

Hasta que indiques que quieres reescribir `df`, lo que has escrito no será permanente. Si quieres que lo reordene y que `df` se mantenga ordenado de esa forma, debes escribir lo siguiente:

````python
df = df.sort_values('Calificación')
df.head() #Este lo escribimos para ver cómo ha quedado.
````

Si quieres ordenar de forma descendente (de mayor a menor) y que el cambio sea permanente, debes aplicar lo siguiente:

````python
df = df.sort_values('Calificación', ascending= False)
df.head() #Este lo escribimos para ver cómo ha quedado.
````

Para cambiar el nombre de las columnas, puedes usar el siguiente código:

````python
df = df.rename(columns= {'Calificación': 'Nota'})
df.head() #Este lo escribimos para ver cómo ha quedado.
````

Para eliminar una columna por completo, se usará:

````python
df = df.drop(columns=['Estatura'])
df.head() #Este lo escribimos para ver cómo ha quedado.
````

#### **_Group by_**

Un **group by** te permite ordenar la información en una serie de condiciones, así como aplicar alguna función que te ayude bastante a comprender un **data frame**.

Ejemplo de lo anterior es que puedes pedirle a tu **data frame**, llamado `df`, que te diga la máxima calificación que encuentre por edad.

````python
df.groupby(by= 'Edad')['Nota'].max()
````

Ahora se explorará cada elemento de este código:
- Primero se indica el **data frame** que se va a revisar, en este caso, `df`, y se agrega `.groupby(`.
- Indica en cuál columna se hará el análisis y cierra el paréntesis (`)`).
- Entre corchetes (`[`, `]`), indica la columna a evaluar; en otras palabras, a cuál le buscará el valor máximo, el promedio, el conteo o lo que tú pidas.
- Por último, indica qué quieres evaluar. Puede ser uno de los siguientes: `.max()` para el máximo valor en una serie de valores; `.min()` para el mínimo valor en una serie de valores; `.count()` para indicar cuántos valores hay en una lista; `.mean()` para conocer el valor promedio en una serie de valores, y `.describe()` para proporcionar un resumen completo de una serie de datos.
